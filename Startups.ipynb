```python
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.get_dummies(df,columns=['State'])
```

```python
le=LabelEncoder()
df['Profit']=le.fit_transform(df['Profit'])
df
```

### here inplace delete the selected row since here inplace run twice it doesnot get the state_newyork as it is deleted permanently in the first phase

```python
df.drop('State_New York',axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop('Profit',axis=1).values,df['Profit'],test_size=0.33,shuffle=True,random_state=0)
```

```python
y_train
```

```python
y_test
```

```python
X_train
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
regressor.score(pd.DataFrame(X_test),pd.DataFrame(y_test))
```

```python
pd.DataFrame(data=list(zip(y_test,y_pred)),columns=['y','y_hat'])
```
